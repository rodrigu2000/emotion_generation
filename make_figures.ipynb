```python
import matplotlib.pyplot as plt
```

```python
persons = ["RR", "hamza", "sarko"]
losses = ["_without_emotion_loss", ""]
emotions = ["happy", "sad", "angry", "surprised"]

n_cols = 5

plt.figure(figsize=(8, 10))

for i_person, person in enumerate(persons):
    for i_loss, loss in enumerate(losses):
        plt.subplot(len(persons) * 2, n_cols, i_person * n_cols * 2 + i_loss * n_cols + 1)
        plt.imshow(plt.imread(f"outputs/inverted_neutral_{person}.jpg"))
        plt.axis("off")
        for i_emotion, emotion in enumerate(emotions):
            plt.subplot(len(persons) * 2, n_cols, i_person * n_cols * 2 + i_loss * n_cols + i_emotion + 2)
            if i_loss == 0:
                plt.imshow(plt.imread(f"outputs/{emotion}_{person}.jpg"))
            else:
                plt.imshow(plt.imread(f"output_Hamza/{person}_neutral_to_{emotion}.jpg"))
            plt.axis("off")
plt.subplots_adjust(wspace=0.1, hspace=0.1)  # Adjust spacing
plt.show()
```

```python
RR = [
    "outputs/input_neutral_RR.jpg",
    "outputs/happy_RR.jpg",
    "output_Hamza/RR_neutral_to_sad.jpg",
    "output_Hamza/RR_neutral_to_angry.jpg",
    "outputs/surprised_RR.jpg"
]

hamza = [
    "outputs/input_neutral_hamza.jpg",
    "outputs/happy_hamza.jpg",
    "output_Hamza/hamza_neutral_to_sad.jpg",
    "output_Hamza/hamza_neutral_to_angry.jpg",
    "outputs/surprised_hamza.jpg"
]

sarko = [
    "outputs/input_neutral_sarko.jpg",
    "outputs/happy_sarko.jpg",
    "output_Hamza/sarko_neutral_to_sad.jpg",
    "output_Hamza/sarko_neutral_to_angry.jpg",
    "output_Hamza/sarko_neutral_to_surprised.jpg"
]

kamala = [
    "outputs/input_neutral_kamala.jpg",
    "outputs/happy_kamala.jpg",
    "outputs/sad_kamala.jpg",
    "outputs/angry_kamala.jpg",
    "outputs/surprised_kamala.jpg"
]

persons = [RR, hamza, sarko, kamala]

plt.figure(figsize=(15, 12))
for i_person, person in enumerate(persons):
    for i, img in enumerate(person):
        plt.subplot(len(persons), 5, i_person * 5 + i + 1)
        plt.imshow(plt.imread(img))
        plt.axis("off")
plt.subplots_adjust(wspace=0.1, hspace=0.1)
plt.show()
```
